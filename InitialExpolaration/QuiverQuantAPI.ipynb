```python
import yfinance as yf
import pandas as pd
import quiverquant
from pathlib import Path
```

```python
token = 'ae7272b6f0ef826789d145ccba3004b2e2e00720'
qq = quiverquant.quiver(token)
```

```python
contracts_df = qq.gov_contracts()
```

```python
df = qq.congress_trading()
df = df[df['TransactionDate'] >= '2023-01-01']
df
```

```python
df.info()
```

```python
alist = []

for index, row in df.iterrows():
    
    try:
        startdate = row[2]
        startprice = yf.download(row[3], start = startdate, end = startdate + pd.Timedelta(days=1))
        endprice = yf.download(row[3], start = '2023-10-12', end = '2023-10-13')
        
        startprice = float(startprice['Close'])
        endprice = float(endprice['Close'])
        
        pct_change = ((endprice - startprice) / startprice) * 100
        
        alist.append(pct_change)

        
    except:
        alist.append('Skipped')
        
df['return_after_trade'] = alist

```

```python
df['return_after_trade'].replace(to_replace = 'Skipped', value=0, inplace=True)
df[df['Transaction'] == 'Sale'].sort_values('return_after_trade', ascending = True)
```
